```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import thompson as tp
```

```python
url = "https://raw.githubusercontent.com/digipodium/Datasets/main/reinforcement/Ads_CTR_Optimisation.csv"
df = pd.read_csv(url)
df.head()
```

```python
# with the alogrithm

result = tp.UCB_random(df)
tp.plot(result)
```

```python
# with ucb alogrithm
result = tp.UCB(df)
tp.plot(result)
```

```python
# with thompson sampling
result = tp.thompson(df)
tp.plot(result)
```
