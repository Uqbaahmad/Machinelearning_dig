#### automobile dataset preprocessing

```python
automoblie_data_url = "https://raw.githubusercontent.com/digipodium/Datasets/main/regression/automobile.csv"
df = pd.read_csv(automoblie_data_url, index_col = 0)
df
```

```python
df.describe()
```

```python
df.info()
```

```python
numerical_col = ['normalized-losses','wheel-base', 'length', 'width', 'height','curb-weight','engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower',
'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

categorical_col = ['symboling', 'make', 'fuel-type','aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'num-of-cylinders', 'fuel-system']
```

```python
# remove all? from dataframe
df.replace('?',np.nan,inplace = True)
```

```python
# numerical column with missing values
df[numerical_col].isnull().sum()
```

```python
df[categorical_col].isnull().sum()
```

```python
num_col_with_na = ['normalized-losses', 'bore', 'stroke', 'horsepower','peak-rpm']
cat_col_with_na = ['num-of-doors']
```

```python
from sklearn.impute import SimpleImputer
```

```python
num_imp = SimpleImputer()
cat_imp = SimpleImputer(strategy='most_frequent')
```

```python
df[num_col_with_na] = num_imp.fit_transform(df[num_col_with_na])
```

```python
df[cat_col_with_na] = cat_imp.fit_transform(df[cat_col_with_na])
```

```python
df[categorical_col].nunique()
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
fuel_enc = LabelEncoder()
df['fuel-type'] = fuel_enc.fit_transform(df['fuel-type'])
df['fuel-type']
```

```python
fuel_enc.classes_
```

```python
fuel_enc.transform(['gas', 'gas', 'diesel'])
```

```python
fuel_enc.inverse_transform([0,1,0])
```

```python

df['aspiration'] = fuel_enc.fit_transform(df['aspiration'])
df['aspiration']
```

```python
fuel_enc.classes_
```

```python
binary_cols = ['fuel-type', 'aspiration', 'num-of-doors', 'engine-location']
ordinal_enc = OrdinalEncoder()
df[binary_cols]= ordinal_enc.fit_transform(df[binary_cols])
df[categorical_col]
```

```python
make_enc = OneHotEncoder(drop='first')
make_dummies = make_enc.fit_transform(df[['make']]).toarray()
```

```python

df = pd.concat([df, pd.DataFrame(make_dummies)], axis=1)
df.head()
```

```python
body_enc = OneHotEncoder(drop = 'first')
body_dummies = body_enc.fit_transform(df[['body-style']]).toarray()
df.drop(columns=['body-style'],inplace=True)
df = pd.concat([df, pd.DataFrame(body_dummies)],axis=1)
```

```python
df.head()
```

```python
engine_enc = OneHotEncoder(drop= 'first')
engine_dummies = engine_enc.fit_transform(df[['engine-type']]).toarray()
df.drop(columns=['engine-type'], inplace=True)
df = pd.concat([df, pd.DataFrame(engine_dummies)],axis=1) 
```

```python
df.head()
```

```python
fuel_enc = OneHotEncoder(drop='first')
fuel_dummies = fuel_enc.fit_transform(df[['fuel-system']]).toarray()
df.drop(columns=['fuel-system'],inplace=True)
df = pd.concat([df, pd.DataFrame(fuel_dummies)],axis=1)
```

```python
df.head()
```

```python
num_enc = OneHotEncoder(drop='first')
num_dummies = num_enc.fit_transform(df[['num-of-cylinders']]).toarray()
df.drop(columns=['num-of-cylinders'],inplace=True)
```
